```python
## Amazon last year stats
import requests
import bs4
from bs4 import BeautifulSoup 

### Downloading the price data

page = requests.get("https://finance.yahoo.com/quote/AMZN/history?p=AMZN")
soup = BeautifulSoup(page.content,"html.parser")
last_year_table = soup.find_all('table',class_= "W(100%) M(0)")
last_year_table
```

```python
## check the table since find.all returns all tables which havs the specific class
len(last_year_table)
```

```python
## check the type of table (find.all returns a result set)
type(last_year_table)
```

```python
## extract one table
last_year_table = last_year_table[0]
```

```python
## what are we looking for 
type(last_year_table)
```

```python
## Tr represents the rows and Td represents the cells
## Use for loop to pull out each cell then found all the rows
## Find all the rows 
## for each row w e want to put on a cell or an element each of the stats
for row in last_year_table.find_all('tr'): 
    for cell in row.find_all('td'): 
        print(cell.text)
```

```python
## change the format to what we want (print to a text file)
with open ("Amazon_last_year_stats_table.txt","w") as stats_file:
    header =("{a:^14}{b:^19}{c:^22}{d:^20}{e:^19}{f:^21}{g:^20}".format(a='Data',b='Open',c='High',d='Low',e='Close*',f='Adj Close**',g='Volume'))
    stats_file.write(header)
    for row in last_year_table.find_all("tr"): 
        for cell in row.find_all("td"): 
            stats_file.write(cell.text.ljust(20))
        stats_file.write('\n')
## not easy to read so we will write all the text of each cell and write a new line 
## devide each row by a new line
## Everything is sort of bunched up together so we need to create some space between each of these freed cell
```
